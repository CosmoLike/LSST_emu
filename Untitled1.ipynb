```python
import numpy as np
from chainconsumer import ChainConsumer


def twochain_single_plot(filename, out, chainnames, paranames):
	d1 = np.genfromtxt(filename[0])
	d2 = np.genfromtxt(filename[1])
	d1new=d1[500000:800000,(3,4)]
	d2new=d2[500000:800000,(3,4)]
	c = ChainConsumer()
	c.add_chain(d1new, parameters=paranames, name =chainnames[0])
	c.add_chain(d2new, name =chainnames[1])
	c.configure(kde=[2.0,2.0],colors=['b','r'], sigmas=[1,2],shade=[True,True],shade_alpha=[0.2,0.2],linestyles=["--", "-"],shade_gradient=0.0)	

	#c.configure(shade=True, shade_alpha=0.2, bar_shade=True)
	#c.configure(shade=[True,False,False],shade_alpha=[0.2,0.2,0.2],linestyles=["--", "-", "-."],linewidths=[0.5,1.,1.])	
	fig = c.plotter.plot(figsize=2.0,filename="/Users/timeifler/Dropbox/cosmolike_store/castor/plots/"+out)
	



filename=["/Users/timeifler/Dropbox/cosmolike_store/castor/like/chain_castorEuclid_3x2pt","/Users/timeifler/Dropbox/cosmolike_store/castor/like/chain_castorLSST_3x2pt"]
chainnames=[r"Castor+Euclid",r"Castor+LSST"]
paranames=[r"$w_0$", r"$w_a$"]
twochain_single_plot(filename,"castor.pdf",chainnames,paranames)
```
